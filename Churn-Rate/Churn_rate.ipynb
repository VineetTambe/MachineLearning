```python
dataset = dataset.drop(columns=['app_web_user'])
```

```python
 # data preparation
user_identfier = dataset['user']
dataset = dataset.drop(columns =['user'])
```

```python
# one Hot encoding
# converting binary categorical into their own individual columns

dataset.housing.value_counts()
```

```python
dataset = pd.get_dummies(dataset)
```

```python
dataset.columns
```

```python
dataset = dataset.drop(columns = ['housing_na','zodiac_sign_na','payment_type_na'])
```

```python
#splitting the dataset into training and testing sets

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataset.drop(columns='churn'),
                                                 dataset['churn'],
                                                 test_size = 0.2,
                                                 random_state=0)
```

```python
y_train.value_counts()
```

```python
pos_index = y_train[y_train == 1].index
neg_index = y_train[y_train == 0].index
```

```python
if len(pos_index)>len(neg_index):
    higher = pos_index
    lower = neg_index
else :
    higher = neg_index
    lower = pos_index

import random
random.seed(0)
higher = np.random.choice(higher,size=len(lower))
lower = np.asarray(lower)

new_indexes = np.concatenate((lower,higher))
```

```python
X_train = X_train.loc[new_indexes]
y_train = y_train[new_indexes]
```

```python
# feature scalling
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
```

```python
X_train_scaled = pd.DataFrame(sc_x.fit_transform(X_train))
X_test_scaled = pd.DataFrame(sc_x.fit_transform(X_test))
```

```python
X_train_scaled.columns = X_train.columns.values
X_test_scaled.columns = X_test.columns.values
```

```python
X_train_scaled.index = X_train.index.values
X_test_scaled.index = X_test.index.values
```

```python
X_train = X_train_scaled
X_test = X_test_scaled
```

```python
X_train.head()
```

```python
# Fitting Model to the Training Set

from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state =0)
classifier.fit(X_train,y_train)
```

```python
# Predicing Test Set
y_pred=classifier.predict(X_test)
```

```python
# Evaluating Result
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score,recall_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
sns.heatmap(cm,annot = True,fmt = 'g')
```

```python
# applying k-Fold cross Validation
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator = classifier,
                            X= X_train,
                            y=y_train,
                            cv =10)
```
